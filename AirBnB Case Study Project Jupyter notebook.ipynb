```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../files/AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.groupby('neighbourhood_group').count()
#There are 5 neigborhood groups. Manhattan shows up the most.
```

```python
# Are private rooms the most popular in manhattan?

air_bnb.groupby(['neighbourhood_group', 'room_type']).count()
#Entire home/apt is the most popular in Man hattan.
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb.groupby('host_name').sum().sort_values('reviews_per_month')
#David, Michael, Alex, Sonder, and John are the busiest.
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby('neighbourhood_group').mean()['price']
# Manhattan
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby('neighbourhood_group').sum()['price']
# Manhattan
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby('host_name').sum().sort_values('price')['price']
# Sonder, Blueground, Michael, David, Alex
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

air_bnb[air_bnb['availability_365'] == 0][air_bnb['number_of_reviews'] >= 100]
```

```python
# What host has the highest total of prices and where are they located?

air_bnb.groupby('host_name').sum().sort_values('price')['price']
air_bnb[air_bnb['host_name'] == 'Sonder (NYC)'].groupby('neighbourhood_group').sum()
# Sonder (NYC). Manhattan
```

```python
# When did Danielle from Queens last receive a review?
air_bnb[air_bnb['host_name'] == 'Danielle'][air_bnb['neighbourhood_group'] == 'Queens']
# 2019-07-08
```

## Further Questions

```python
air_bnb.groupby('host_name').sum().sort_values('id')['id']
# Sonder (NYC)
```
